###### Thay thế giá trị khuyết thiếu

```python
# Thay thế giá trị khuyết thiếu của thuộc tính Description bằng "Unknown"
retail["Description"] = retail["Description"].fillna("Unknown")
```

```python
# Thay thế giá trị khuyết thiếu của thuộc tính CustomerID bằng giá trị phía trước
retail["CustomerID"] = retail["CustomerID"].fillna(method = "ffill")
```

###### Xác định giá trị ngoại lai của thuộc tính Quantity và Thuộc tính UnitPrice

```python
# Giá trị ngoại lai của Quantity là những giá trị < 0
# Giá trị ngoại lai của UnitPrice là những giá trị <= 0
retail_ = retail[~((retail["Quantity"] <= 0) & (retail["UnitPrice"] <= 0))]
```
