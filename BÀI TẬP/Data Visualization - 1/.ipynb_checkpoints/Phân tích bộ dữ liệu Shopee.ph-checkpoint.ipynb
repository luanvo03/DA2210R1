```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
shopee = pd.read_csv("D:/DA2210R1/data/shopeep_koreantop_clothing_shop_data.csv")
```

```python
shopee.head()
```

```python
shopee.info()
```

```python
shopee["shop_location"] = shopee["shop_location"].fillna("Unknown")
shopee = shopee.dropna()
```

### Vẽ biểu đồ so sánh số lượng shop gia nhập theo các năm.

```python
shops = shopee["join_year"].value_counts()
shops = shops.sort_index()
```

```python
plt.bar(shops.index, shops.values)
plt.title("The number of shops joined by years", fontsize = 16, color = "r")
plt.xlabel("Years")
plt.ylabel("The number of shops joined")
plt.show()
```

### Vẽ biểu đồ thể hiện mối quan hệ giữa tỉ lệ phản hồi với số lượt khách hàng đánh giá tốt.

```python
plt.scatter(shopee["response_rate"], shopee["rating_good"])
plt.title("Relationship between response rate and the number of good customer reviews", color = "r")
plt.xlabel("Response rate")
plt.ylabel("The number of good customer reviews")
plt.show()
```

### Vẽ biểu đồ thể hiện mối quan hệ giữa thời gian phản hồi (đơn vị giây) với số lượt khách hàng đánh giá xấu.

```python
plt.scatter(shopee["response_time"], shopee["rating_bad"])
plt.title("Relationship between response time and the number of bad customer reviews", color = "r")
plt.xlabel("Response time")
plt.ylabel("The number of bad customer reviews")
plt.show()
```

### Vẽ biểu đồ thể hiện xu hướng của số lượng shop gia nhập theo thời gian.

```python
plt.plot(shops.index, shops.values)
plt.title("The trend of the number of shops joining over time", fontsize = 16, color = "r")
plt.xlabel("Years")
plt.ylabel("The number of shops joined")
plt.show()
```

### Vẽ biểu đồ thể hiện phân bố của điểm đánh giá trung bình.

```python
plt.hist(shopee["rating_star"])
plt.title("Distribution of average rating", fontsize = 16, color = "r")
plt.xlabel("Average rating")
plt.ylabel("Number of rating")
plt.show()
```
