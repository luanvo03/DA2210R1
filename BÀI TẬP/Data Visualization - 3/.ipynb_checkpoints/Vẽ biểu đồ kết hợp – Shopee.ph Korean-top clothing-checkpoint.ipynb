```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
shopee = pd.read_csv("D:/DA2210R1/data/shopeep_koreantop_clothing_shop_data.csv")
```

```python
shopee.info()
```

```python
shopee.head()
```

### Vẽ biểu đồ tần số số lượng shop gia nhập theo các năm.

```python
sns.countplot(x = "join_year", data = shopee)
```

### Vẽ biểu đồ xu hướng thể hiện mối quan hệ giữa tỉ lệ phản hồi với số lượt khách hàng đánh giá tốt.

```python
sns.lmplot(x = "response_rate", y = "rating_good", data = shopee)
```

### Vẽ biểu đồ thể hiện phân bố của điểm đánh giá trung bình. 

```python
sns.violinplot(y = "rating_star", data = shopee)
```

### Vẽ biểu đồ tần số của cửa hàng chính thức và không chính thức. 

```python
sns.countplot(x = "is_official_shop", data = shopee)
```

### Vẽ biểu đồ tần số của cửa hàng được xác thực với chưa xác thực. 

```python
sns.countplot(x = "is_shopee_verified", data = shopee)
```
