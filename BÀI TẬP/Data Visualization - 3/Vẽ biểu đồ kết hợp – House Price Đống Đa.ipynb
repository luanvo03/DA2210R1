```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
house_price = pd.read_csv("D:/DA2210R1/data/house_price_Dống-Da_Hà-Nội_subdata.csv")
```

```python
house_price.info()
```

```python
house_price.head()
```

### Vẽ biểu đồ xu hướng phân tích mối quan hệ giữa giá nhà với diện tích, giá nhà với số lượng phòng ngủ, nhận xét.

```python
sns.lmplot(x = "area", y = "price", data = house_price)
sns.lmplot(x = "bedroom", y = "price", data = house_price)
plt.show()
```

### Vẽ biểu đồ phân bố thể hiện phân bố của giá nhà theo các hướng, nhận xét.

```python
sns.violinplot(y = "price", x = "house_direction", data = house_price)
```

### Vẽ biểu đồ tần số để đếm số nhà ở mỗi hướng nhà, nhận xét.

```python
sns.countplot(x = "house_direction", data = house_price)
```

### Vẽ biểu đồ boxplot thể hiện phân bố của giá nhà theo các hướng, nhận xét.

```python
sns.boxplot(x = "house_direction", y = "price", data = house_price)
```
