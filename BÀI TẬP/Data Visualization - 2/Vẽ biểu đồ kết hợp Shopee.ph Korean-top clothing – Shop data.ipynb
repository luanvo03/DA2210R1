```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
shopee = pd.read_csv("D:/DA2210R1/data/shopeep_koreantop_clothing_shop_data.csv")
```

```python
shopee.info()
```

### Vẽ biểu đồ:
So sánh số lượng shop gia nhập theo các năm.

Xu hướng của số lượng shop gia nhập theo từng tháng trong từng năm.

```python
shop_per_year = shopee["join_year"].value_counts().sort_index()
```

```python
shop_per_month = shopee.groupby(["join_year", "join_month"])["join_month"].count().to_frame()
```

```python
shop_per_month.columns = ["no_shops_join"]
```

```python
shop_per_month = shop_per_month.reset_index()
```

```python
shop_per_month["join_date"] = shop_per_month["join_month"].apply(str) + " " + shop_per_month["join_year"].apply(str)
```

```python
shop_per_month
```

```python
fig, ax = plt.subplots(1, 2, figsize = (15, 15))
ax[0].bar(shop_per_year.index, shop_per_year.values)

ax[1].plot(shop_per_month["join_date"], shop_per_month["no_shops_join"])

plt.show()
```
