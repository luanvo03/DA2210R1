```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Biểu đồ để hiển thị giá trị cụ thể và so sánh GDP các nước Vietnam,  Indonesia, Cambodia, Thailand và Malaysia.

### Biểu đồ để đánh giá tỉ lệ đóng góp GDP của các nước trên tổng số GDP của 5 nước Vietnam,  Indonesia, Cambodia, Thailand và Malaysia.

```python
gdp = pd.read_csv("D:/DA2210R1/data/GDPlist.csv", encoding = "ISO-8859-1")
```

```python
gdp.head()
```

```python
gdp.info()
```

```python
countries = ["\xa0Vietnam", "\xa0Indonesia", "\xa0Cambodia", "\xa0Thailand", "\xa0Malaysia"]
```

```python
countries_5 = gdp[gdp["Country"].isin(countries)]
countries_5
```

```python
fig, ax = plt.subplots(1, 2)
ax[0].bar(countries_5["Country"], countries_5["GDP (millions of US$)"])
ax[0].set_title("GDP Các Nước")
ax[0].set_xlabel("Countries")
ax[0].set_ylabel("GDP")

ax[1].pie(countries_5["GDP (millions of US$)"], labels = countries_5["Country"].values, autopct = "%1.2f%%")

plt.show()
```
