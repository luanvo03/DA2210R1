### Hợp nhất các DataFrame

```python
df1 = pd.DataFrame({'employee': ['Bob', 'Jake', 'Lisa', 'Sue'],'group': ['Accounting', 'Engineering', 'Engineering', 'HR']})
df2 = pd.DataFrame({'employee': ['Lisa', 'Bob', 'Jake', 'Sue'],'hire_date': [2004, 2008, 2012, 2014]})
print(df1); 
print(df2)
```

```python
df3 = pd.merge(df1, df2)
df3
```

```python
pd.merge(df1, df2, on = 'employee')
```

```python
df1
```

```python
df3 = pd.DataFrame({'name': ['Bob', 'Jake', 'Lisa', 'Sue'],'salary': [70000, 80000, 120000, 90000]})
df3
```

```python
pd.merge(df1, df3, left_on  = 'employee', right_on = 'name')
```

```python
# Concat
ser1 = pd.Series(["A", "B", "C"], index = [1, 2, 3])
ser2 = pd.Series(["D", "E", "F"], index = [4, 5, 6])
pd.concat(ser1, ser2)
```

### Groupby and Pivot table

```python
data = pd.read_csv("D:/DA2210R1/data/students.csv")
```

```python
data.head()
```

```python
data.groupby("Quê quán")
```

```python
data.groupby("Quê quán")["Toán"]
```

```python
data.groupby("Quê quán")[["Toán", "Văn"]].mean()
```

```python
data.groupby("Quê quán")["Mã học sinh"].count()
```

```python
data.head()
```

```python
data.groupby(["Quê quán", "Giới tính"])["Toán"].aggregate("mean").unstack()
```

```python
data.groupby(["Quê quán", "Giới tính"])["Toán"].aggregate("mean")
```

```python
data.groupby(["Quê quán", "Giới tính"])["Toán"].aggregate("mean").to_frame()
```

```python
data.pivot_table("Toán", index = "Quê quán", columns = "Giới tính")
```
