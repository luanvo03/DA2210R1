#### Lọc bằng phương thức filter

```python
# Lọc các cột
student.filter(["Toán", "Lý", "Hoá"])
```

#### Lọc có điều kiện

```python
# Lọc theo điều kiện đơn
student[(student["Toán"]>7)]
```

```python
# Lọc theo nhiều điều kiện
# & - và, | - hoặc, ~ - phủ định
student[(student["Toán"]>7) & (student["Văn"]>=6.5)]
```

```python
student[~((student["Toán"]>7) & (student["Văn"]>= 6.5))]
```

```python
# Lọc có điều kiện bằng phương thức query
student.query("Toán >7 and Văn >=6.5")
```

```python
student.query("Toán >7 and not Văn >=6.5")
```
