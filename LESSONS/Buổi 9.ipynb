```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
population = pd.read_csv("D:/DA2210R1/data/population_book0.csv", index_col = "Country")
```

```python
population.head()
```

```python
population.info()
```

```python
population = population.fillna(population.mean())
population
```

```python
from sklearn.preprocessing import StandardScaler
s_scaler = StandardScaler()
population_StandardScaler = s_scaler.fit_transform(population)
population_StandardScaler = pd.DataFrame(population_StandardScaler)
```

```python
sns.boxplot(x = population_StandardScaler[0])
```

```python
plt.boxplot(x = population_StandardScaler[0])
```

```python
from sklearn.preprocessing import MinMaxScaler
m_scaler = MinMaxScaler()
population_MinMaxScaler = m_scaler.fit_transform(population)
population_MinMaxScaler = pd.DataFrame(population_MinMaxScaler)
```

```python
sns.boxplot(x = population_MinMaxScaler[0])
```

```python
plt.boxplot(population_MinMaxScaler[0])
```

```python
from sklearn.preprocessing import RobustScaler
r_scaler = RobustScaler()
population_RobustScaler = r_scaler.fit_transform(population)
population_RobustScaler = pd.DataFrame(population_RobustScaler)
```

```python
sns.boxplot(x = population_RobustScaler[0])
```

```python
plt.boxplot(x = population_RobustScaler[0])
```
