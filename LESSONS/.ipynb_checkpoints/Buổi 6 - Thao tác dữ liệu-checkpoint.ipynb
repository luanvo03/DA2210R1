```python
import pandas as pd
```

## Lọc dữ liệu

```python
student = pd.read_csv("D:/DA2210R1/data/students.csv", encoding = "utf-8")
```

#### Dùng chỉ số cột và dòng để lọc dữ liệu

```python
student["Toán"]
```

```python
# Lọc các cọt dữ liệu, fancy indexing
student[["Toán", "Lý", "Hoá", "Sinh", "Văn"]]
```

```python
student[student.columns[5:]]
```

```python
# Lọc dữ liệu theo dòng: loc and iloc
student[1:5]
```

```python
student.loc[1:5]
```

```python
student.iloc[1:5]
```

```python
# Lọc dữ liệu theo cột và dòng
student[student.columns[5:]][:4]
```

```python
student.iloc[:3, :2]
```

#### Lọc bằng phương thức filter

```python
# Lọc các cột
student.filter(["Toán", "Lý", "Hoá"])
```

```python
student.filter(student.columns[5:])
```

```python
# Lọc các dòng rời rạc
student.filter([0, 1, 3, 5, 7], axis = 0)
```

```python
student.filter(student.index[student.index%2!=0], axis =0)
```

```python
# Lọc theo cột và dòng
student.filter(["Toán", "Lý", "Hoá"])[:5]
```

#### Lọc có điều kiện

```python
# Lọc theo điều kiện đơn
student[(student["Toán"]>7)]
```

```python
# Lọc theo nhiều điều kiện
# & - và, | - hoặc, ~ - phủ định
student[(student["Toán"]>7) & (student["Văn"]>=6.5)]
```

```python
student[~((student["Toán"]>7) & (student["Văn"]>= 6.5))]
```

```python
# Lọc có điều kiện bằng phương thức query
student.query("Toán >7 and Văn >=6.5")
```

```python
student.query("Toán >7 and not Văn >=6.5")
```

## Gom nhóm dữ liệu

```python
student.groupby("Quê quán")["Toán"].mean()
```
